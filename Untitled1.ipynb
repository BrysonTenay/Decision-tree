```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Example dataset structure
import pandas as pd
data = pd.DataFrame({
    'GPA': [3.2, 3.8, 2.9, 3.5],
    'TestScores': [85, 90, 70, 88],
    'Income': [40000, 55000, 30000, 60000],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'ProgramInterest': ['Engineering', 'Science', 'Arts', 'Science'],
    'Enrolled': [1, 1, 0, 1]
})

# Encode categorical variables
data = pd.get_dummies(data, columns=['Gender', 'ProgramInterest'], drop_first=True)

# Features and target
X = data.drop('Enrolled', axis=1)
y = data['Enrolled']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train decision tree
dt_model = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_model.fit(X_train, y_train)

# Test predictions
y_pred = dt_model.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plot_tree(dt_model, feature_names=X.columns, class_names=['Not Enrolled', 'Enrolled'], filled=True)
plt.show()
```

```python
feature_importances = dt_model.feature_importances_
for feature, importance in zip(X.columns, feature_importances):
    print(f"{feature}: {importance:.2f}")
```

```python
#saving
import joblib
joblib.dump(dt_model, 'decision_tree_model.pkl')
```
